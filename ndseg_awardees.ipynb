Step 1:
Can we find records of grantees in a public publication API like [ORCID's Public API](https://pub.orcid.org/v2.0/#/).

```python
import pandas as pd
from requests_xml import XMLSession
```

```python
# awardee.txt taken from: 
# https://ndseg.asee.org/ndseg_fellows/2014_awardees
# TODO: automate name scraping part
data = pd.read_csv("awardees.txt", sep='\t')
awardees = data["Awardee"].tolist()
```

```python
BASE_URL = "https://pub.orcid.org/v2.0/search?q=%22{}%22"

session = XMLSession()

orcids = dict()

for awardee in awardees:
    name_string = awardee.replace(" ","%20")
    name_string = name_string.replace("-","%20")
    awardee_url = BASE_URL.format(name_string)
    r = session.get(awardee_url)
    results = r.xml.search("<common:path>{}</common:path>")
    awardee_orcids = list()
    if len(results) >0:
        for result in results:
            awardee_orcids.append(result[0])
    orcids[awardee] = awardee_orcids
```
